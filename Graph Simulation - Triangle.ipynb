```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

N = 3 # side length of triangle
G = nx.generators.lattice.triangular_lattice_graph(N, 2*N-1, with_positions=True)
pos = nx.get_node_attributes(G, 'pos')
```

```python
# Generate Triangle Graph (stupid method lol)

acceptable_nodes = []

# Add the index of all nodes that are in the triangle graph
for i in range(N+1):
    if i < N//2:
        acceptable_nodes.extend([(i,j) for j in range(2*i+2)])
    elif i == N//2:
        acceptable_nodes.extend([(i,j) for j in range(N+1)])
    elif i > N//2:
        acceptable_nodes.extend([(i,j) for j in range(2*(N-i)+1)])
        
remove_nodes = [n for n in G.nodes if n not in acceptable_nodes]

G.remove_nodes_from(remove_nodes)
```

```python
fig = nx.draw(G, with_labels = True, node_size = 1000, pos=pos)
plt.show()
```
