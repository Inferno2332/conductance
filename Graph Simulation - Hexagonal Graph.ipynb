```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

N = 4  # number of rows of the lattice structure
G = nx.generators.lattice.hexagonal_lattice_graph(N, 2*N-1)
pos = nx.get_node_attributes(G, 'pos')
```

```python
'''
plot the figure 
'''
fig = nx.draw(G, with_labels = True, node_size = 100, pos=pos)
plt.show()
```
